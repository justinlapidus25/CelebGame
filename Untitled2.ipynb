```python
import pandas as pd
import re

with open("celebs.txt.rtf", "r", encoding="utf-8", errors="ignore") as f:
    text = f.read()

# Remove basic RTF formatting
text = re.sub(r"{\\.*?}", "", text)
text = re.sub(r"\\[a-z]+\d* ?", "", text)
text = text.replace("{", "").replace("}", "")

names = text.split(",")

cleaned_names = []

for name in names:
    name = name.strip()
    name = re.sub(r"\s+", " ", name)

    if len(name.split()) >= 2:
        cleaned_names.append(name)

df = pd.DataFrame(cleaned_names, columns=["Name"])

df = df.drop_duplicates()

df.to_csv("my_celebs.csv", index=False)

print(df.head())
print(len(df))
```
